```python
#라이브러리 import
import numpy as np
import pandas as pd
```

```python
uv_108 = pd.read_csv("../processed uv data/uv_stn108.csv")
uv_108.head(5)
```

```python
#날짜, 시간 데이터 -> object 형으로 변환
uv_108['yyyymmdd'] = uv_108['yyyymmdd'].astype(str)
uv_108['hhnn'] = uv_108['hhnn'].astype(str)

uv_108.info()
```

```python
#날짜 형식: yyyy-mm-dd
#시간 형식: hh:nn
uv_108['yyyymmdd'] = pd.to_datetime(uv_108['yyyymmdd'])
uv_108['yyyymmdd'] = uv_108['yyyymmdd'].astype(str)

uv_108.head(5)
```

```python
for i in range(len(uv_108)):
    new_hhnn = '{0:0>4}'.format(uv_108['hhnn'][i])
    hhnn = '{0}:{1}'.format(new_hhnn[0:2], new_hhnn[2:])
    uv_108['hhnn'][i] = hhnn
```

```python
uv_108.head(5)
```

```python
#yyyymmdd와 hhnn 합쳐서 새로운 datetime형 feature 만들기
uv_108['dateTime'] = uv_108['yyyymmdd'] + " " + uv_108['hhnn']
uv_108.head(5)
```

```python
uv_108['dateTime'] = pd.to_datetime(uv_108['dateTime'])
uv_108 = uv_108.drop(['yyyymmdd', 'hhnn', 'lon', 'lat'], axis=1)
uv_108.head(5)
```

```python
uv108_data = uv_108[(uv_108['dateTime']>= '2020-05-01') & (uv_108['dateTime']< '2020-07-01')]
uv108_data.head(5)
```

```python
uv108_data.tail(5)
```

```python
uv108_data.to_csv('./data/uv108.csv', index=False)
```

```python
#테스트 데이터 불러오기
test = uv_108[(uv_108['dateTime']>= '2021-05-01') & (uv_108['dateTime']< '2021-07-01')]
test.head(5)
```

```python
#X, Y 지정
columns = ['band1', 'band2', 'band3', 'band4', 'band5', 'band6', 'band7', 'band8', 'band9', 'band10', 'band11',
                'band12', 'band13', 'band14', 'band15', 'band16', 'solarza', 'sateza', 'esr']
X = uv108_data[columns]
Y = uv108_data['uv']

#train set : test set ~=~ 0.8 : 0.2
test_data = test.sample(frac = 0.25)
x_test = test_data[columns]
y_test = test_data['uv']
```

```python
from sklearn import linear_model

lr = linear_model.LinearRegression()
model = lr.fit(X, Y)
```

```python
#회귀 계수
coef1 = pd.Series(data = lr.coef_, index = X.columns)
coef1
```

```python
#모델 test
y_predict = lr.predict(x_test)
```

```python
#모델 평가
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse)

print('rmse: ', rmse)
```

```python
#유의성 검정
import statsmodels.api as sm
X2 = sm.add_constant(X)

model2 = sm.OLS(Y, X2).fit()
model2.summary()
```
