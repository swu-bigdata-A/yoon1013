```python
#라이브러리 import
import numpy as np
import pandas as pd
```

```python
#data set: stn 13
uv_13 = pd.read_csv("../processed uv data/uv_stn13.csv")
uv_13.head(5)
```

```python
print('dataset 구조: ', uv_13.shape)
```

• band9: 중층 수승기 밴드  
• band10: 하층 수증기 밴드  
• band11: 구름상 밴드  
• band12: 오존 밴드  
• band13: 대기창 밴드  
• band14: 깨끗한 대기창 밴드  
• band15: 오염된 대기창 밴드  
• band16: 이산화탄소 밴드  
• solarza: 태양 천정각  
• esr: 대기외 일사량

```python
#날짜, 시간 데이터 -> object 형으로 변환
uv_13['yyyymmdd'] = uv_13['yyyymmdd'].astype(str)
uv_13['hhnn'] = uv_13['hhnn'].astype(str)

uv_13.info()
```

```python
#날짜 형식: yyyy-mm-dd
#시간 형식: hh:nn
uv_13['yyyymmdd'] = pd.to_datetime(uv_13['yyyymmdd'])
uv_13['yyyymmdd'] = uv_13['yyyymmdd'].astype(str)
```

```python
uv_13.head(5)
```

```python
for i in range(len(uv_13)):
    new_hhnn = '{0:0>4}'.format(uv_13['hhnn'][i])
    hhnn = '{0}:{1}'.format(new_hhnn[0:2], new_hhnn[2:])
    uv_13['hhnn'][i] = hhnn
```

```python
uv_13.head(5)
```

```python
#yyyymmdd와 hhnn 합쳐서 새로운 datetime형 feature 만들기
uv_13['dateTime'] = uv_13['yyyymmdd'] + " " + uv_13['hhnn']
```

```python
uv_13.head(5)
```

```python
uv_13['dateTime'] = pd.to_datetime(uv_13['dateTime'])
```

```python
uv_13 = uv_13.drop(['yyyymmdd', 'hhnn', 'lon', 'lat'], axis=1)
```

```python
uv_13.head(5)
```

```python
uv13_data = uv_13[(uv_13['dateTime']>= '2020-05-01') & (uv_13['dateTime']< '2020-07-01')]
```

```python
uv13_data.head(5)
```

```python
uv13_data.tail(5)
```

```python
#데이터 표준화
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scale_columns = ['band1', 'band2', 'band3', 'band4', 'band5', 'band6', 'band7', 'band8', 'band9', 'band10', 'band11',
                'band12', 'band13', 'band14', 'band15', 'band16', 'solarza', 'sateza', 'esr']
uv13_standard_data = uv13_data
uv13_standard_data[scale_columns] = scaler.fit_transform(uv13_standard_data[scale_columns])
```

```python
uv13_standard_data.head(5)
```

```python
uv13_data.to_csv('./data/uv13.csv', index=False)
uv13_standard_data.to_csv('./data/uv13_standard.csv', index=False)
```

```python
#테스트 데이터 불러오기
test = uv_13[(uv_13['dateTime']>= '2021-05-01') & (uv_13['dateTime']< '2021-07-01')]
test.head(5)
```

```python
#테스트 데이터 정규화
test_standard = test
test_standard[scale_columns] = scaler.fit_transform(test_standard[scale_columns])
```

```python
#X, Y 지정
columns = ['band1', 'band2', 'band3', 'band4', 'band5', 'band6', 'band7', 'band8', 'band9', 'band10', 'band11',
                'band12', 'band13', 'band14', 'band15', 'band16', 'solarza', 'sateza', 'esr']
X = uv13_data[columns]
Y = uv13_data['uv']

#train set : test set ~=~ 0.8 : 0.2
test_data = test.sample(frac = 0.25)
x_test = test_data[columns]
y_test = test_data['uv']
```

```python
#linear regression - VIF 고려안함
from sklearn import linear_model

lr = linear_model.LinearRegression()
model = lr.fit(X, Y)
```

```python
#회귀 계수
coef1 = pd.Series(data = lr.coef_, index = X.columns)
coef1
```

```python
#모델 test
y_predict = lr.predict(x_test)
```

```python
#모델 평가
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse)

print('rmse: ', rmse)
```

```python
#유의성 검정
import statsmodels.api as sm
X2 = sm.add_constant(X)

model2 = sm.OLS(Y, X2).fit()
model2.summary()
```
